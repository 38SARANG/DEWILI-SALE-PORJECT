```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# plotting a bar chart for Gender and it's count

ax = sns.countplot(x ='Gender',data = df,hue='Gender')
sns.set(rc={'figure.figsize':(7,3)})

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
ax = sns.lineplot(x='Age', y='Amount', data=df, hue='Gender',style='Gender')
sns.set(rc={'figure.figsize':(12,7)})


```


From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men

# **AGE**

```python
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```


# From above graphs we can see that most of the buyers are of age group between 26-35 yrs female

# **STATE**

```python
# total number of orders from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(5)
sales_state

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'State',y= 'Orders')
```

# From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka

# Marital Status

```python
Ax = sns.countplot(data = df, x = 'Marital_Status',hue='Gender')

sns.set(rc={'figure.figsize':(10,3)})
for bars in ax.containers:
    ax.bar_label(bars)
plt.title('MALE                 FEMALE')
plt.xlabel('Marital Status')
plt.ylabel('Count')
plt.show()
```

# Occupation

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Occupation',hue='Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

# From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Product_Category', hue='Product_Category', palette='rocket')

for bars in ax.containers:
    ax.bar_label(bars),
    plt.xticks(rotation=90)
```
